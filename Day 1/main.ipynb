```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["Species"] =iris.target

#show first rows
print(df.head())

#Shape of the data
print("Shape : ",df.shape)

#Dataset info
print(df.info())

#Statistical summary
print(df.describe())

```
